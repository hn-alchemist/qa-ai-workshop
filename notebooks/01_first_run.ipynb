### Step 4 — A helper to ask the model anything
An LLM takes **messages** (a system role that sets behaviour + your user message)
and returns text. This `ask()` wraps that.

```python
def ask(question, system='You are a helpful software QA expert.', max_new_tokens=400):
    messages = [{'role': 'system', 'content': system},
                {'role': 'user', 'content': question}]
    out = chat(messages, max_new_tokens=max_new_tokens, do_sample=False)
    return out[0]['generated_text'][-1]['content']

print(ask('In one paragraph, explain the difference between a smoke test and a regression test.'))
```
