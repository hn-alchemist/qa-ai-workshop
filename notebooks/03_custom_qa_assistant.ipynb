# 🧪 03 · Build Your Own QA-Bot (no training needed)

**Level:** Juniors and up · **Time:** ~20 min

Before *training* a model (next notebook), learn the cheaper, faster way to specialise one:
a strong **system prompt**. You'll turn a general model into *your team's* QA assistant —
with your standards, format, and tone baked in.

> ⚙️ Enable GPU: *Runtime → Change runtime type → T4 GPU*.

```python
!pip install -q -U transformers accelerate
import torch
from transformers import pipeline
chat = pipeline('text-generation', model='Qwen/Qwen2.5-1.5B-Instruct',
                torch_dtype=torch.bfloat16, device_map='auto')
print('✅ ready')
```

### A weak vs. a strong system prompt
Same model, same question — only the *persona* changes. Watch the output quality jump.

```python
def reply(system, user, max_new_tokens=400):
    out = chat([{'role':'system','content':system}, {'role':'user','content':user}],
               max_new_tokens=max_new_tokens, do_sample=False)
    return out[0]['generated_text'][-1]['content']

question = 'Review this bug report: \"App crashes sometimes.\"'

weak = 'You are a helpful assistant.'
print('--- WEAK PROMPT ---')
print(reply(weak, question))
```

```python
# A strong, opinionated QA persona — THIS is your 'custom bot'
QA_BOT = '''You are 'QA-Bot', a meticulous senior QA reviewer for our team.
Rules you always follow:
- Never accept vague bug reports. Demand: clear title, exact steps to reproduce,
  expected vs actual result, environment, and severity.
- When a report is weak, REWRITE it into our gold-standard format and point out what was missing.
- Be concise and use checklists.'''

print('--- QA-BOT ---')
print(reply(QA_BOT, question))
```

### Your turn 🛠️
Edit `QA_BOT` to match *your* team's real standards (your bug template, your severity scale,
your definition of done). This persona text is reusable everywhere — in notebooks, in the
Hugging Face Space app, or as an Ollama Modelfile if you ever go local.

```python
MY_QA_BOT = '''You are our team's QA assistant. (Edit me!)
Replace these lines with your real standards...'''

print(reply(MY_QA_BOT, 'Write acceptance criteria for a file-upload feature (max 10MB, images only).'))
```

---
### 🧠 Lesson
- **90% of 'custom AI' is just a great prompt** — fast, free, no training.
- Train (next notebook) only when a prompt can't capture the behaviour, or you have lots of
  examples of your team's exact style you want the model to *absorb*.
